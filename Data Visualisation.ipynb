```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
# available in notebooks only
#%matplotlib inline
```

```python
x = np.random.rand(10)
x
```

```python
plt.plot(x)
# x-axis is just indices
# y-axis is x
```

```python
# styling
style.use('ggplot') # grid architecture
plt.plot(x,'g',label='first line graph', linewidth=5)
plt.xlabel('index')
plt.ylabel('random values')
plt.title('my first graph')
plt.show() # header line goes away
```

```python
# create time series plot
time = np.linspace(0,23,24)
# x and y must have same first dimension i.e 24
values = np.random.randint(100,1000,24)
plt.plot(time,values,'y',linewidth=3)
plt.xlabel('hours')
plt.ylabel('traffic')
plt.title('time-series graph')
plt.show()
```

```python
# changing the axis
# plt.legend()
plt.plot(time,values,'y',label='web traffic',linewidth=3)
plt.axis([0,25,100,1000])
plt.xlabel('hours')
plt.ylabel('traffic')
plt.title('time-series graph')
plt.legend()
plt.show()
```

```python
# multiplot
sun = np.random.randint(100,1000,24)
sat = np.random.randint(100,1000,24)
plt.plot(time,sun,'y',label='sun',linewidth=1)
plt.plot(time,sat,'b',label='sat',linewidth=1)
plt.title('Multiple Plots')
plt.legend()
plt.show()
```

```python
# Subplots
fig = plt.figure()
plt.subplot(2,1,1)
plt.plot(time,sun,'y',label='sun',linewidth=1)
plt.legend()
plt.title('Subplots')
plt.subplot(2,1,2)
plt.plot(time,sat,'b',label='sat',linewidth=1)

plt.legend()
plt.show()
```

```python
fig = plt.figure()
plt.subplot(1,2,2)
plt.plot(time,sun,'y',label='sun',linewidth=1)
plt.legend()
plt.subplot(1,2,1)
plt.plot(time,sat,'b',label='sat',linewidth=1)
plt.legend()
plt.show()
```
